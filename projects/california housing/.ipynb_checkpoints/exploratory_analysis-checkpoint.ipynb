```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_parquet('california_housing.parquet')
target = ['med_houseval']
dataset.head()
```

```python
dataset.info(verbose=True)
```

```python
ax = pd.plotting.scatter_matrix(dataset, alpha=0.1, figsize=(30,10))
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(dataset.corr('spearman'), annot=True, ax = ax)
fig.suptitle('Spearman Correlation')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(dataset.corr('pearson'), annot=True, ax = ax)
fig.suptitle('Pearson Correlation')
```

```python
dataset.plot.scatter(x='longitude', y='latitude', alpha=0.2, figsize=(5,5), title='Longitude x Latitude in California')
```
